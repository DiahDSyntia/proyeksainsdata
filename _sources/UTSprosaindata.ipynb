## Load Data

```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
import re
import swifter
df = pd.read_csv("/content/drive/MyDrive/proyeksaindata/tugas/dataabstrak.csv")
text = df["Abstrak"]
text
```

## Case Folding

```python
def case_folding(comment):
    comment = comment.lower()
    return comment
text = text.apply(case_folding)
text
```

## Menghilangkan karakter, link, dan Stopword yang ada

```python
def linkNormalize(text):
    text = re.sub(r"\s—\s", "", text)
    text = re.sub(r"http\S+", "", text)
    return text
text = text.apply(linkNormalize)
text
```

```python
symbols = ",!\"#$%&()*+-.…/:;<=>?@[\]^_`{|}~\n0987654321"
def cleaning_text(comment):
    for i in symbols:
        comment = np.char.replace(comment, i, ' ')
    return comment
text = text.apply(cleaning_text)
text
```

```python
nltk.download('stopwords')
list_stopwords = stopwords.words('indonesian')
def stopword_removal(comment):
    comment = comment.split()
    data = [word for word in comment if word not in list_stopwords]
    kalimat = ""
    for i in data:
      kalimat += i
      kalimat += " "
    return kalimat
```

## Memperbarui kata yang tidak baku menggunakan Sastrawi

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()
def stemmingNormalize(comment):
    comment = comment.split()
    data = []
    for term in comment:
        term = stemmer.stem(term)
        data.append(term)
    kalimat = ""
    for i in data:
      kalimat += i
      kalimat += " "
    return kalimat
text = text.swifter.apply(stemmingNormalize)
text
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_separate = tfidf_vectorizer.fit_transform(text)

df_tfidf = pd.DataFrame(
    tfidf_separate.toarray(), columns=tfidf_vectorizer.get_feature_names_out(), index=df.index
)
X = df_tfidf.values
df_tfidf
```

## Reduksi Dimensi (PCA)

```python
from sklearn.decomposition import PCA

pca = PCA(n_components =10)

X = pca.fit_transform(X)
# X_test = pca.transform(X_test)

pca.explained_variance_ratio_
```

## K-MEAN CLUSTERING

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)
```

```python
center = kmeans.cluster_centers_
center
```

```python
y_pred = kmeans.labels_


data = kmeans.labels_
res = []
for i in data:
    if i == 0:
        res.append("RPL")
    else:
        res.append("KOMPUTASI")
print(res)
```

```python
y_true = df['Kategori'].values.tolist()
print(df['Kategori'].values.tolist())
```

```python
n = len(y_true)
for i in range(n):
    print(res[i],"\t : ",y_true[i])
```

### Hasil Akurasi

```python
from sklearn.metrics import accuracy_score
print ("Accuracy : ",
    accuracy_score(y_true, res)*100)
```
