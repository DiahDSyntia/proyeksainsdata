```python
import nltk
from nltk.corpus import stopwords
import re
import swifter
```

```python
text = df['Abstrak']
Y = df['Kategori'].values
text
```

### Case Folding

```python
def case_folding(comment):
    comment = comment.lower()
    return comment
text = text.apply(case_folding)
text
```

### Menghilangkan link yang ada

```python
def linkNormalize(text):
    text = re.sub(r"\s—\s", "", text)
    text = re.sub(r"http\S+", "", text)
    return text
text = text.apply(linkNormalize)
text
```

### Menghilangkan karakter tertentu

```python
symbols = ",!\"#$%&()*+-.…/:;<=>?@[\]^_`{|}~\n"
def cleaning_text(comment):
    for i in symbols:
        comment = np.char.replace(comment, i, ' ')
    return comment
text = text.apply(cleaning_text)
text
```

menggunakan library untuk menghilangkan stopwords yang ada

```python
nltk.download('stopwords')
list_stopwords = stopwords.words('indonesian')
def stopword_removal(comment):
    comment = comment.split()
    data = [word for word in comment if word not in list_stopwords]
    kalimat = ""
    for i in data:
      kalimat += i
      kalimat += " "
    return kalimat
```

```python
text = text.apply(stopword_removal)
```

Memperbarui kata yang tidak baku menggunakan sastrawi

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()
def stemmingNormalize(comment):
    comment = comment.split()
    data = []
    for term in comment:
        term = stemmer.stem(term)
        data.append(term)
    kalimat = ""
    for i in data:
      kalimat += i
      kalimat += " "
    return kalimat
text = text.swifter.apply(stemmingNormalize)
text
```

## TF-IDF

Dibawah ini adalah code untuk TF-IDF.

tfidf_separate = tfidf_vectorizer.fit_transform(text). Dan untuk code tersebut, text diambil dari code sebelum ini, jadi saling terhubung dengan code diatas.

Dan hasilnya seperti dibawah ini

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_separate = tfidf_vectorizer.fit_transform(text)

df_tfidf = pd.DataFrame(
    tfidf_separate.toarray(), columns=tfidf_vectorizer.get_feature_names_out(), index=df.index
)
X = df_tfidf.values
df_tfidf
```

## Data Training Testing
Setelah itu menggunakan code seperti dibawah ini untuk training dan testing. Dengan test size sebanyak 0.2 dan random statenya 100

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( 
    X, Y, test_size = 0.2, random_state = 100)
print("Jumlah Data training : ", len(X_train))
print("Jumlah Data test : ", len(X_test))
```

## Menghitung Akurasi

Menghitung akurasi menggunakan PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 10)

X_train_2 = pca.fit_transform(X_train)
X_test_2 = pca.transform(X_test)

pca.explained_variance_ratio_
```

Import metode, metode yang digunakan saat ini adalah KNN. dibawah ini adalah code untuk metode tersebut

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train_2, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = classifier.predict(X_test_2)
print (" KNN Accuracy : ",
    accuracy_score(y_test,y_pred)*100)
```
